```python
total_projects = int(len(outcome))

top_10 = np.random.randint(0, int(total_projects), int(total_projects * 0.1))
outcome_10 = outcome.iloc[top_10]
not_funded = ((outcome_10["fully_funded"] == 'f').astype(int)).sum()

# percent of not funded
print(not_funded / len(outcome_10))
```

```python
out_df = outcome[["projectid", "fully_funded"]]
data = out_df.join(project_df.set_index('projectid'), on='projectid', how='inner', lsuffix='_left', rsuffix='_right')
data.sort_values(by=['total_price_excluding_optional_support'], inplace=True, ascending=False)
data_10 = data.iloc[np.arange(int(len(data) * 0.1))]
not_funded = ((data_10["fully_funded"] == 'f').astype(int)).sum()
print(not_funded / len(data_10))
```
